<b style="color: red;">1. Introduzione al diabete e al contesto</b> <br>
   Il diabete è una malattia in cui c’è aumento nel sangue dei livelli di glucosio (zucchero; la glicemia) per un deficit della quantità e,     spesso, nell’efficacia biologica dell’insulina, l’ormone che controlla la glicemia nel sangue e che viene prodotto dal pancreas. Il          diabete non è una malattia contagiosa, e non è una malattia ereditaria, nel senso che tranne che per poche varietà molto rare, non c’è un    passaggio inevitabile della malattia da una generazione ad un’altra. Esiste però una predisposizione familiare, soprattutto in caso di       diabete tipo 2, per cui chi ha un diabetico fra i parenti di primo grado (genitori, fratelli) ha un rischio di ammalarsi superiore           rispetto a chi non ha parenti con la malattia. (fonte SID)

   In questo progetto ci occuperemo di prevedere diagnosticamente se un paziente ha il diabete o meno in base a determinate misurazioni         diagnostiche incluse nel dataset utilizzato e attraverso dei modelli di Machine Learning. 

   Il dataset su cui lavoriamo proviene originariamente dal National Institute of Diabetes and Digestive and Kidney Diseases. Sono stati        imposti diversi vincoli alla selezione di queste istanze da un database più ampio. In particolare, tutti i pazienti qui sono donne di        almeno 21 anni di discendenza indiana Pima. I Pima sono gli indiani americani che abitano nella fetta di territorio compreso tra             l’Arizona centrale e meridionale, negli Stati Uniti, e lo stato del Sonora in Messico.

<b style="color: red;">2. Descrizione dei dati </b> <br>
Il dataset è costituito da un insieme di variabili predittive mediche che saranno le nostre features, variabili indipendenti che    influenzano l'output, cioè il target (variabile dipendente), che qui è presente come altro attributo nel dataset con il nome di "Outcome", l'attributo Oucome può assumere due valori: 0 se il paziente non è malato, 1 se il paziente è malato. Dunque il problema è un problema di classificazione binaria.

Le variabili indipendenti includono: <br>
<b> 1. Gravidanze: </b> Il numero di gravidanze che la paziente ha avuto <br>
<b> 2. Glucosio: </b> Concentrazione plasmatica di glucosio a 2 ore in un test di tolleranza al glucosio orale <br>
<b> 3. Pressione sanguigna: </b> Pressione diastolica (mm Hg), che rappresenta la pressione nelle arterie durante il riposo del cuore, cioè              quando non si contrae e pompa il sangue <br>
            Le persone con pressione alta hanno un rischio maggiore di sviluppare il diabete di tipo 2. Questo è dovuto al fatto che                     l'ipertensione e il diabete spesso condividono fattori di rischio comuni, come l'obesità, l'infiammazione cronica, e uno stile               di vita poco salutare. Inoltre, la pressione alta può danneggiare i vasi sanguigni e influire negativamente sulla capacità del               corpo di regolare correttamente il glucosio. <br>
<b> 4. Spessore della pelle: </b> Spessore della piega cutanea del tricipite (mm) <br>
Lo spessore della piega cutanea è un indicatore della percentuale di grasso corporeo. Un accumulo eccessivo di grasso corporeo, specialmente intorno alla vita, è fortemente correlato con il rischio di sviluppare il diabete. <br>
<br> Pressione sanguigna alta e spessore della piega cutanea (indicatore dell'obesità) sono misurazioni che contribuiscono a valutare il rischio di diabete, in quanto legate a uno stato metabolico alterato e alla resistenza all'insulina, condizioni chiave che predispongono al diabete di tipo 2. <br>
<b> 5. Insulina: </b> Insulina sierica a 2 ore (mu U/ml) <br>
si riferisce alla concentrazione di insulina nel siero del sangue, che è la porzione liquida del sangue priva di cellule e proteine di coagulazione. È una misura utilizzata per valutare la funzione del pancreas e il metabolismo del glucosio nel corpo <br>
<b> 6. BMI: </b> ndice di massa corporea (peso in kg/(altezza in m)^2) <br>
è una misura utilizzata per valutare se una persona ha un peso sano in relazione alla sua altezza e è comunemente utilizzato per stimare il rischio di sviluppare malattie correlate al peso, come il diabete di tipo 2 <br>
Sottopeso: BMI inferiore a 18,5 <br>
Normopeso: BMI tra 18,5 e 24,9 <br>
Sovrappeso: BMI tra 25 e 29,9 <br>
Obesità: BMI uguale o superiore a 30 <br>
<b> 7. Diabetes Pedigree Funcion (DPF): </b> Funzione del pedigree del diabete <br>
è una misura utilizzata nel campo della medicina e del machine learning per valutare il rischio ereditario di diabete di una persona, basandosi sulla storia familiare. Insieme a altre variabili è una delle più importanti per la predizione. <br>
<b> 8. Eta': </b> (anni) <br>
In particolare per quanto riguarda diabete di tipo 2, tende a manifestarsi con maggiore frequenza nelle persone più anziane

<b style="color:red;"> Sostituzione dei valori mancanti </b>

Seguendo prima il metodo dell'imputazione ora i valori mancanti (Nan) devono a loro volta essere rimpiazzati per non falsare le predizioni che vedremo successivamente. Questi valori Nan possono essere sostituiti con diverse tecniche, quella che useremo è <b> l'Imputazione condizionale sulla base dell'outcome </b>: Invece di sostituire i valori mancanti in una colonna utilizzando la mediana calcolata su tutti i dati, cioè una volta ordinata la lista si prende il valore in mezzo a questa, la mediana viene calcolata separatamente per ciascun gruppo definito dalla variabile di outcome (o target), dunque non dipende solo dalla posizione, ma anche dai vari insiemi/classi di Outcome.

Qui di seguito troveremo la funzione <b> median_target(var) </b> che serve per calcolare la mediana di una variabile 'var' per ciascun gruppo dell'outcome, in questo caso Outcome 0 = non diabetico, Outcome 1 = diabetico.

```python
def median_target(var):   
    temp = data[data[var].notnull()]   # Creo un sottoinsieme del dataset nel quale non ci sono valori nulli per quella data variabile
    temp = temp[[var, 'Outcome']].groupby(['Outcome'])[[var]].median().reset_index()   # Qui per ciascun gruppo viene calcolata la mediana per la variabile passata come paametro
    return temp
```

<b style="color:red;"> 6.5 Feature Selection </b> <br>
La feature selection è una parte del processo nella quale vengono selezionate delle feature risultate più rilevanti nell'analisi precedente e eliminate alcune ridondanti, che possono introdurre rumore sucessivamente. Sicuramente abbiamo visto come l'Insulina, i livelli di Glucosio e altre variabili abbiano una certa importanza rispetto ad altre <br>
<b> L'univariate selection </b> è una delle tecniche utilizzata nell'ambito del Machine Learning per la selezione delle caratteristiche ai fini di una predizione (feature selection), le metriche di selezione statistiche valutano la relazione tra le caratteristiche e il target
Un'altra tecnica utilizzata per la scelta delle feature migliori è la <b> Recursive Feature Elimination </b> (RFE) che utilizza RFE per eliminare ricorsivamente le feature meno importanti. <br>
Tuttavia avendo solamente 8 feature, potrebbe essere sconveniente toglierne qualcuna dal momento che, non c'è una forte ridondanza tra variabili e si potrebbero perdere informazioni utili alla predizione del diabete. (Riferimento al pairplot e alla matrice di confusione per capire la correlazione tra variabili) <br>
Proveremo comunque a eseguire una feature selection per poi effettuare un confronto e verificare se la precisione risulta essere maggiore utilizzando la suddetta tecnica oppure lasciando il numero di feature invariato.

<b> Ora utilizzeremo la Recursive Feature Elimination per scoprire quali sono le feature considerate più importanti </b> <br>
RFE restituisce un modello addestrato sulle feature selezionate e un array di booleani che indicano quali feature sono state mantenute (ovvero dove compare True).

```python
from sklearn.feature_selection import RFE

# Definizione del modello
model = LogisticRegression()

# Creazione di RFE con selezione di 4 feature
rfe = RFE(estimator=model, n_features_to_select=4)

# Adattamento di RFE al dataset di training
rfe.fit(x_train, y_train)

# Stampa dei risultati
print("Feature selezionate:", rfe.support_)
print("Ranking delle feature:", rfe.ranking_)
```

Scegliendo 4 come il numero di feature più importanti per determinare la malattia il risultato è: <b> Numero di gravidanze, Glucosio, BMI e DPF </b>

```python
# Selezione delle feature più importanti
selected_features = x_train.columns[rfe.support_]
print("Feature selezionate:", selected_features)
```

```python
# Creazione dei dataset ridotti per addestramento, validazione e test
x_train_reduced = x_train[selected_features]
x_val_reduced = x_val[selected_features]
x_test_reduced = x_test[selected_features]
```
